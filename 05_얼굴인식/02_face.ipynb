```python
import cv2
import matplotlib.pyplot as plt
import cvlib as cv               # 파이썬에서 얼굴, 객체 인식을 위한 사용하기 쉬운 라이브러리이다.
import numpy as np
```

```python
image_path = 'IVE.jpg'        # 이미지 경로
im = cv2.imread(image_path) # 이미지 읽기
plt.imshow(im)              # 이미지 보여주기 
plt.show()
```

```python
# detect faces (얼굴 검출)
faces, confidences = cv.detect_face(im)  #  얼굴의 좌표들과 예측된 얼굴 각각의 confidence
```

```python
for face in faces:
    (startX,startY) = face[0],face[1]  # 좌표 추출 
    (endX,endY) = face[2],face[3]
    # draw rectangle over face
    cv2.rectangle(im, (startX,startY), (endX,endY), (0,255,0), 2)  # 색상과 두께 

# display output        
plt.imshow(im)
plt.show()
cv2.imwrite('result.jpg', im) # 이미지 쓰기
```
